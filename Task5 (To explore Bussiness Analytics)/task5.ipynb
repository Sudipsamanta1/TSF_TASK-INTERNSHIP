# Import Library

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
%matplotlib inline
```

# Visualization

```python
plt.figure(figsize=(20,9))
sns.heatmap(correlation,annot=True)
plt.show()
```

```python
data['Ship Mode'].value_counts()
```

```python
plt.figure(figsize=(18,9))
sns.countplot(x=data['Ship Mode'])
plt.show()
```

```python
data['City'].value_counts()
```

```python
plt.figure(figsize=(18,6))
sns.countplot(x=data['City'],order=(data["City"].value_counts().head(10)).index)
plt.show()
```

```python
print(data["State"].value_counts())

plt.figure(figsize=(20,10))
sns.countplot(x=data["State"].head(100))
plt.show()

plt.figure(figsize=(20,10))
sns.countplot(x=data["State"].tail(100))
plt.show()
```

```python
plt.figure(figsize=(18,12))
plt.pie(data['Sub-Category'].value_counts(),labels=data['Sub-Category'].value_counts().index,startangle=180,radius=1)
plt.show()
```

```python
plt.figure(figsize=(20,7))
sns.barplot(x=data["Sub-Category"],y=data["Profit"])
plt.show()
```
